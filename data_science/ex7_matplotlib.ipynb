```python
""" Sine curve Plot """
import numpy as np
import matplotlib.pyplot as plt
plot_path = '/workspaces/docker_python/data_science/plots'
x = np.linspace(0,10,1000) # x is a 1000 evenly spaced numbers from 0 to 10 (Sampling kind of thing)
y = np.sin(x)
fig = plt.figure()
plt.plot(x, y)
plt.savefig(plot_path + "/plot.png")
plt.show()
```

```python
""" Labels and Titles """
import numpy as np
import matplotlib.pyplot as plt
x = np.linspace(0,10,1000) # 1darray of length 1000
y = np.sin(x)

plt.plot(x, y)
plt.xlabel('x')
plt.ylabel('y')
plt.title('function sin(x)')
plt.savefig(plot_path + "/plot.png")
plt.show()
```

```python
""" Multiple Lines """
import numpy as np
import matplotlib.pyplot as plt
x = np.linspace(0,10,1000) # 1darray of length 1000
y = np.sin(x)
plt.plot(x, np.sin(x))
plt.plot(x, np.cos(x))
plt.savefig(plot_path + "/plot.png")
plt.show()

x = np.linspace(0,10,1000) # 1darray of length 1000
y = np.sin(x)
plt.plot(x, np.sin(x), color='k')
plt.plot(x, np.cos(x), color='r', linestyle ='--')
plt.savefig(plot_path + "/plot.png")
plt.show()
```

```python
""" Legend """
# create a plot legend labeling each line when Multiple Lines 
import numpy as np
import matplotlib.pyplot as plt
x = np.linspace(0,10,1000) # 1darray of length 1000
y = np.sin(x)
plt.plot(x, np.sin(x), 'k:', label='sin(x)')
plt.plot(x, np.cos(x), 'r--', label='cos(x)')
plt.legend()
plt.savefig(plot_path + "/plot.png")
plt.show()
```

```python
""" Scatter Plot """
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd

presidents_df = pd.read_csv('https://sololearn.com/uploads/files/president_heights_party.csv', index_col='name')
plt.scatter(presidents_df['height'], presidents_df['age'])
plt.savefig(plot_path + "/plot.png")
plt.show()

plt.scatter(presidents_df['height'], presidents_df['age'],
   marker='<',
   color='b')
plt.xlabel('height'); 
plt.ylabel('age')
plt.title('U.S. presidents')
plt.savefig(plot_path + "/plot.png")
plt.show()
```

```python
""" Plotting with Pandas """
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd

presidents_df = pd.read_csv('https://sololearn.com/uploads/files/president_heights_party.csv', index_col='name')
presidents_df.plot(kind = 'scatter', 
  x = 'height', 
  y = 'age',
  title = 'U.S. presidents')
plt.savefig(plot_path + "/plot.png")
plt.show()
```

```python
""" Histogram
A histogram shows the underlying frequency distribution of a set of continuous data,
 allowing for inspection of the data for its shape, outliers, skewness, etc. 
"""
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd

#Plot directly from pandas dataframe
presidents_df = pd.read_csv('https://sololearn.com/uploads/files/president_heights_party.csv', index_col='name')
presidents_df['height'].plot(kind='hist',
  title = 'height',
  bins=5)
plt.savefig(plot_path + "/plot.png")
plt.show()

# using plt functionality
plt.hist(presidents_df['height'], bins=5)
plt.savefig(plot_path + "/plot.png")
plt.show()
```

```python
""" Box Plot """
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd

presidents_df = pd.read_csv('https://sololearn.com/uploads/files/president_heights_party.csv', index_col='name')
plt.style.use('classic')
presidents_df.boxplot(column='height');
plt.savefig(plot_path + "/plot.png")
plt.show()
```

```python
""" Bar Plot """
# Histogram presents numerical data whereas bar plot shows categorical data
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd

presidents_df = pd.read_csv('https://sololearn.com/uploads/files/president_heights_party.csv', index_col='name')
party_cnt = presidents_df['party'].value_counts()

plt.style.use('ggplot')
party_cnt.plot(kind ='bar')
plt.savefig(plot_path + "/plot.png")
plt.show()
```
